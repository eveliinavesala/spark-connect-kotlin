```kotlin
import org.apache.spark.sql.SparkSession
import pragmatic.toDataFrame
import pragmatic.toKotlinList
```

```kotlin
val spark = SparkSession.builder()
    .remote("sc://localhost:15002")
    .getOrCreate()
```

```kotlin
data class Person(val name: String, val age: Int)

val people = listOf(
    Person("Alice", 30),
    Person("Bob", 25),
    Person("Charlie", 35)
)

val df = people.toDataFrame(spark)
```

```kotlin
df.show()
```

```kotlin
val olderPeopleDF = df.filter(df.col("age").gt(28))

val olderPeople: List<Person> = olderPeopleDF.toKotlinList()

olderPeople
```

```kotlin
spark.stop()
```
